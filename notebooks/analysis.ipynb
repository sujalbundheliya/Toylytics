```python
import pandas as pd
df = pd.read_csv("../data/cleaned_data.csv")

top_brands = df["Brand"].value_counts().head(5)
top_rated_brands = df.groupby("Brand")["Rating"].mean().sort_values(ascending=False).head(5)

# Plot brand frequency
top_brands.plot(kind="bar", title="Top 5 Brands by Frequency")

# Pie chart of brand share
top_brands.plot(kind="pie", autopct="%1.1f%%", title="Brand Share of Top 5")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
sns.scatterplot(data=df, x="Rating", y="Selling Price")

# Average price by rating bin
df["Rating Bin"] = pd.cut(df["Rating"], bins=[0, 2, 3, 4, 5], labels=["0-2", "2-3", "3-4", "4-5"])
df.groupby("Rating Bin")["Selling Price"].mean().plot(kind="bar", title="Avg Price by Rating")
```

```python
top_reviewed = df.sort_values("Reviews", ascending=False).head(5)
top_rated = df.sort_values("Rating", ascending=False).head(5)

# Top-reviewed bar chart
top_reviewed.plot.bar(x="Title", y="Reviews", title="Top 5 Most Reviewed")

# Top-rated bar chart
top_rated.plot.bar(x="Title", y="Rating", title="Top 5 Highest Rated")
```
